```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.linspace(0.001, 0.999, 500)

# Case 1
loss1 = -np.log(1-z)

# Case 2
loss2 = -np.log(z)

plt.figure(figsize=(10, 6))
plt.plot(z, loss1, label='Case 1', color='red')
plt.plot(z, loss2, label='Case 2', color='blue')
plt.legend(fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.xlabel('Probability z', fontsize=12)
plt.xlim(0, 1)
plt.ylim(0, 5)
plt.show
```
